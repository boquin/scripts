```python
import os
os.chdir("/Users/macproajb")
from ipynb.fs.full.FXpricelist import full_df

df = 1 / full_df(20,0)
emlist = ['BRL', 'MXN', 'CLP', 'ZAR', 'TRY', 'PLN', 'HUF', 'CZK', 
          'CNY', 'KRW','SGD', 'MYR', 'IDR', 'INR', 'PHP', 'THB','COP']
list_length = len(emlist)
df = df[emlist]
items = df.columns
days = 100

pairs = []
for i in range(len(items)):
    for j in range(i + 1, len(items)):
        pair = (items[i], items[j])
        pairs.append(pair)
```

```python
corr_list = pd.DataFrame()

for i in range(0,len(pairs)):
    corr_list[i] = df[pairs[i][0]].rolling(days).corr(df[pairs[i][1]])
    print(i, end='')
    
corr_list["mean"] = corr_list.mean(axis=1)
corr_list["90th"] = corr_list["mean"].quantile(0.9)
corr_list["10th"] = corr_list["mean"].quantile(0.1)

num_rows_to_add = 200
# empty_df = pd.DataFrame(np.nan, index=range(num_rows_to_add), columns=corr_list.columns)
# corr_list = corr_list.append(empty_df, ignore_index=True)


em_perf = np.sign(df[emlist].diff(days)).sum(axis=1) / list_length * 100
# empty_series = pd.Series([np.nan] * num_rows_to_add)
# em_perf = em_perf.append(empty_series, ignore_index=True)
```

```python
from plotly.subplots import make_subplots
fig = make_subplots(rows=2, cols=1, shared_xaxes=True,)

fig.add_scatter(x=corr_list.index, y=corr_list["mean"], mode='lines', row=1, col=1,showlegend=False)
fig.add_scatter(x=corr_list.index, y=corr_list["90th"], mode='lines', row=1, col=1,showlegend=False)
fig.add_scatter(x=corr_list.index, y=corr_list["10th"], mode='lines', row=1, col=1,showlegend=False)
fig.add_scatter(y=em_perf, x=em_perf.index, mode='lines', row=2, col=1,showlegend=False)

fig.update_yaxes(title_text="Average Pairwise Correlation", row=1, col=1)
fig.update_yaxes(title_text=f"% of EM pairs up/down in last {days} days", row=2, col=1)
fig.update_layout(width=1000, height=800)

fig.show()
```
