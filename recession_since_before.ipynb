```python
from ipynb.fs.full.FXpricelist import full_df
from ipynb.fs.full.quick_functions import *

df = get_fred("USRECDM",50)
df.columns = ['recession']
df['since'] = 0
df['before'] = 0
for i in range(len(df)):
    if df.iloc[i, 0] == 1:
        df.iloc[i, 1] = 0
    else:
        df.iloc[i, 1] = df.iloc[i-1, 1] + 1 if i > 0 else 1
df2 = df.iloc[::-1]
for i in range(len(df2)):
    if df2.iloc[i, 0] == 1:
        df2.iloc[i, 2] = 0
    else:
        df2.iloc[i, 2] = df2.iloc[i-1, 2] + 1 if i > 0 else 1
df["before"] = df2["before"]
df.plot()
```
